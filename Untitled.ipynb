```python
from google import genai

client = genai.Client()

interaction = client.interactions.create(
    model="gemini-2.5-flash",
    input="hi my name is riya"
)

print(interaction.output_text)
```

```python
def generate_recipe(ingridients, cuisine, diet):
    '''
    generate recipe using ingredient provided.
    cuisine can be "Italian", "Mexican", "Indian", "any" etc.
    diet can be "vegan", "vegetarian", "jain", "any" etc.
    '''
    prompt = f'''
    Generate one food recipe using these ingridients: {','.join(ingridients)}
    Recipe should not be more than 100 words 
    Cuisine: {cuisine}
    Diet:{diet}
    '''
    response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents = prompt
)

    return response.text

recipe = generate_recipe(["tomatoes", "onions"], "Indian", "Vegan")
recipe
```
